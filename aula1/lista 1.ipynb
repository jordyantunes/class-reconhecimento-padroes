```python
import numpy as np
```

```python
m_covariancia = np.matrix([[1.2,0.4],[0.4,1.8]])
```

```python
m_covariancia
```

```python
c1 = np.array([0.5,0.5])
c2 = np.array([2.0,2.0])
c3 = np.array([1.3,1.8])
```

```python
a1 = np.random.multivariate_normal(c1,m_covariancia,300)

a2 = np.random.multivariate_normal(c2,m_covariancia,300)

a3 = np.random.multivariate_normal(c3,m_covariancia,300)
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
g = [a1,a2,a3]
for i in g:
    plt.scatter(i[:,0],i[:,1],alpha=0.8)
```
